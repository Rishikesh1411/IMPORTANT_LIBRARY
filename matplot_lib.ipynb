```python
# Import the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Define the data points for the x-axis and y-axis
x = [1, 2, 3, 4, 5]  # X-axis values
y = [2, 3, 5, 7, 11]  # Y-axis values

# Create a line plot with circular markers at each data point
plt.plot(x, y, marker='o')

# Set the title of the plot
plt.title('Line Plot')

# Label the x-axis
plt.xlabel('X-axis')

# Label the y-axis
plt.ylabel('Y-axis')

# Display the plot
plt.show()

```

```python
# Define the data points for the x-axis and y-axis
x = [5, 7, 8, 7, 2, 17, 2, 9]  # X-axis values
y = [99, 86, 87, 88, 100, 86, 103, 87]  # Y-axis values

# Create a scatter plot with red-colored points
plt.scatter(x, y, color='red')

# Set the title of the plot
plt.title('Scatter Plot')

# Label the x-axis
plt.xlabel('X')

# Label the y-axis
plt.ylabel('Y')

# Display the scatter plot
plt.show()
```

```python
# Define categories for the x-axis
categories = ['A', 'B', 'C', 'D']

# Define corresponding values for each category (bar heights)
values = [10, 24, 36, 40]

# Create a bar plot using Matplotlib
plt.bar(categories, values)

# Set the title of the plot
plt.title('Bar Plot')

# Label the x-axis
plt.xlabel('Category')

# Label the y-axis
plt.ylabel('Values')

# Display the bar plot
plt.show()
```

```python
# Create a horizontal bar chart using Matplotlib
plt.barh(categories, values, color='skyblue')

# Set the title of the plot
plt.title('Horizontal Bar Plot')

# Label the x-axis
plt.xlabel('Values')

# Label the y-axis
plt.ylabel('Category')

# Display the plot
plt.show()
```

```python
# Import NumPy for numerical computations
import numpy as np

# Generate a dataset of 1000 random numbers following a normal distribution
data = np.random.randn(1000)

# Create a histogram with 30 bins
plt.hist(data, bins=30, color='green', edgecolor='black')

# Set the title of the plot
plt.title('Histogram')

# Label the x-axis
plt.xlabel('Value')

# Label the y-axis
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```

```python
# Define the sizes of each section in the pie chart
sizes = [30, 25, 20, 25]

# Define labels for each section
labels = ['A', 'B', 'C', 'D']

# Create a pie chart using Matplotlib
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Set the title of the chart
plt.title('Pie Chart')

# Display the pie chart
plt.show()
```

```python
# Generate three groups of random data using normal distribution
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a box plot using Matplotlib
plt.boxplot(data, vert=True, patch_artist=True)

# Set the title of the plot
plt.title('Box Plot')

# Label the x-axis
plt.xlabel('Group')

# Label the y-axis
plt.ylabel('Values')

# Display the box plot
plt.show()
```

```python
# Create a violin plot using Matplotlib
plt.violinplot(data)

# Set the title of the plot
plt.title('Violin Plot')

# Label the x-axis
plt.xlabel('Group')

# Label the y-axis
plt.ylabel('Values')

# Display the violin plot
plt.show()
```

```python
# Generate a 5x5 matrix of random values between 0 and 1
matrix = np.random.rand(5, 5)

# Display the matrix as a heatmap using Matplotlib
plt.imshow(matrix, cmap='hot', interpolation='nearest')

# Add a color bar to show the scale of values
plt.colorbar()

# Set the title of the heatmap
plt.title('Heatmap')

# Display the heatmap
plt.show()
```

```python
# Generate x values ranging from 0 to 9 with a step of 1
x = np.arange(0, 10, 1)

# Calculate y values using an exponential decay function
y = np.exp(-x / 3)

# Define error values for each y point
errors = 0.1 + 0.2 * x

# Create an error bar plot with black circular markers ('o')
plt.errorbar(x, y, yerr=errors, fmt='o', color='black')

# Set the title of the plot
plt.title('Error Bar Plot')

# Label the x-axis
plt.xlabel('X')

# Label the y-axis
plt.ylabel('Y')

# Display the error bar plot
plt.show()
```

```python
# Define the x-axis values representing days
days = [1, 2, 3, 4, 5]

# Define the quantity of apples and bananas sold or consumed over these days
apples = [3, 2, 4, 3, 4]  # Apple data series
bananas = [1, 2, 1, 2, 1]  # Banana data series

# Create a stack plot to visualize cumulative values over time
plt.stackplot(days, apples, bananas, labels=['Apples', 'Bananas'])

# Add a legend in the upper left corner for clarity
plt.legend(loc='upper left')

# Set the title of the plot
plt.title('Stack Plot')

# Label the x-axis
plt.xlabel('Day')

# Label the y-axis
plt.ylabel('Quantity')

# Display the stack plot
plt.show()
```

```python
# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Plot a line chart in the top-left subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line')

# Plot a bar chart in the top-right subplot
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar')

# Plot a histogram in the bottom-left subplot using the first dataset
axs[1, 0].hist(data[0], bins=10)
axs[1, 0].set_title('Histogram')

# Plot a scatter plot in the bottom-right subplot
axs[1, 1].scatter(x, y)
axs[1, 1].set_title('Scatter')

# Adjust spacing between subplots for better layout
plt.tight_layout()

# Display the complete figure with all four subplots
plt.show()
```
