```python
# Import Seaborn and Matplotlib for data visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

# Create a scatter plot showing the relationship between total bill amount and tip amount
sns.scatterplot(x="total_bill", y="tip", data=tips)

# Display the plot
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", size="size", sizes=(50, 200), data=tips)
plt.show()
```

```python
# Load the built-in "fmri" dataset from Seaborn
fmri = sns.load_dataset("fmri")

# Create a line plot to visualize signal change over time
sns.lineplot(x="timepoint", y="signal", data=fmri)

# Display the plot
plt.show()
```

```python
sns.lineplot(
    x="timepoint",    # X-axis represents different time points in the fMRI experiment
    y="signal",       # Y-axis shows the measured signal strength
    hue="event",      # Different colors are used to represent various events in the experiment
    style="region",   # Line styles differentiate between brain regions
    markers=True,     # Enables markers to highlight individual data points
    dashes=False,     # Disables dashed lines, making the plot easier to read
    data=fmri         # Uses the built-in "fmri" dataset
)

plt.show()  # Displays the line plot
```

```python
# Load the built-in "titanic" dataset from Seaborn
titanic = sns.load_dataset("titanic")

# Create a bar plot to visualize average fare across different passenger classes
sns.barplot(x="class", y="fare", data=titanic)

# Display the plot
plt.show()
```

```python
sns.barplot(
    x="class",     # X-axis represents passenger class (First, Second, Third)
    y="fare",      # Y-axis represents the average fare paid
    hue="sex",     # Bars are color-coded by sex (Male, Female)
    ci=None,       # Disables confidence intervals to show exact averages
    palette="muted",  # Uses a soft, visually pleasing color palette
    data=titanic   # Uses the Titanic dataset from Seaborn
)

plt.show()  # Displays the bar plot
```

```python
# Load the built-in "iris" dataset from Seaborn
iris = sns.load_dataset("iris")

# Create a histogram to visualize the distribution of petal lengths
sns.histplot(x="petal_length", data=iris)

# Display the histogram
plt.show()

```

```python
sns.histplot(
    data=iris,       # Use the Iris dataset
    x="petal_length", # X-axis represents the petal length
    bins=20,         # Divide data into 20 bins for better granularity
    kde=True,        # Add a Kernel Density Estimate (KDE) curve for smooth distribution
    color="green"    # Set histogram color to green
)

plt.show()  # Display the histogram
```

```python
# Load the built-in "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

# Create a box plot to visualize total bill amounts across different days
sns.boxplot(x="day", y="total_bill", data=tips)

# Display the plot
plt.show()
```

```python
sns.boxplot(
    x="day",        # X-axis represents the day of the week
    y="total_bill", # Y-axis shows the total bill amount
    hue="time",     # Color differentiation by meal type (Lunch or Dinner)
    data=tips,      # Uses the built-in "tips" dataset from Seaborn
    palette="Set3"  # Uses the 'Set3' color palette for clarity
)

plt.show()  # Displays the box plot
```

```python
# Create a violin plot using Seaborn to analyze petal length distribution across species
sns.violinplot(x="species", y="petal_length", data=iris)

# Display the plot
plt.show()
```

```python
# Load the built-in "tips" dataset from Seaborn
tips = sns.load_dataset('tips')

# Select only numerical columns before calculating the correlation matrix
# Alternatively, you could explicitly list the numerical columns:
# numerical_tips = tips[['total_bill', 'tip', 'size']]
corr = tips._get_numeric_data().corr()

# Generate a heatmap to visualize correlations between features
sns.heatmap(corr)

# Display the heatmap
plt.show()
```

```python
# Load the built-in "flights" dataset from Seaborn
flights = sns.load_dataset('flights')

# Reshape the dataset into a pivot table where months are rows, years are columns, and values represent the number of passengers
# Explicitly use keyword arguments for the pivot method
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Create a heatmap to visualize the number of passengers per month and year
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d')

# Display the heatmap
plt.show()
```

```python
# Load the built-in "iris" dataset from Seaborn
iris = sns.load_dataset("iris")

# Create a pair plot to visualize relationships between different numerical features
sns.pairplot(data=iris)

# Display the plot
plt.show()
```

```python
sns.pairplot(
    data=iris,      # Use the Iris dataset
    hue="species",  # Color-code data points by species
    diag_kind="kde", # Use Kernel Density Estimation for diagonal plots
    palette="husl"  # Apply the 'husl' color palette for better differentiation
)

plt.show()  # Display the pair plot
```

```python
sns.jointplot(
    x="sepal_length",  # X-axis represents the sepal length
    y="sepal_width",   # Y-axis represents the sepal width
    data=iris          # Use the Iris dataset
)

plt.show()  # Display the joint plot
```

```python
sns.jointplot(
    data=iris,        # Use the Iris dataset
    x="sepal_length", # X-axis represents sepal length
    y="sepal_width",  # Y-axis represents sepal width
    kind="kde",       # Use Kernel Density Estimation (KDE) for smoother visualization
    hue="species"     # Color-code data points by species
)

plt.show()  # Display the joint plot
```

```python
sns.catplot(
    data=tips,      # Use the tips dataset
    kind="swarm",   # Create a swarm plot to visualize individual data points
    x="day",        # X-axis represents the day of the week
    y="total_bill", # Y-axis represents the total bill amount
    hue="smoker"    # Color-code data points by smoking status
)

plt.show()  # Display the plot
```

```python
sns.lmplot(
    data=tips,      # Use the tips dataset
    x="total_bill", # X-axis represents total bill amount
    y="tip",        # Y-axis represents tip amount
    col="time",     # Creates separate plots for Lunch and Dinner
    hue="smoker"    # Color-codes data points based on smoking status
)

plt.show()  # Display the plot
```

```python
sns.displot(
    data=tips,      # Use the tips dataset
    x="total_bill", # X-axis represents total bill amount
    col="time",     # Creates separate plots for Lunch and Dinner
    kde=True        # Adds a Kernel Density Estimation (KDE) curve for smooth distribution visualization
)

plt.show()  # Display the distribution plot
```
