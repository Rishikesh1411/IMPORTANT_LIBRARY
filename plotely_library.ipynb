```python
# Import the Plotly Express library for creating interactive visualizations
import plotly.express as px

# Create a bar chart using Plotly Express
# 'x' contains the labels for the x-axis (categories)
# 'y' contains the values (heights of bars) for each category
fig = px.bar(x=["a", "b", "c"], y=[1, 3, 2])

# Display the bar chart in an interactive window
fig.show()
```

```python
# Import the Plotly Express library for creating interactive visualizations
import plotly.express as px

# Create a line chart using Plotly Express
# 'x' contains the labels for the x-axis (categories)
# 'y' contains the values (data points) corresponding to each category
# 'title' sets the title of the figure
fig = px.line(x=["a", "b", "c"], y=[1, 3, 2], title="sample figure")

# Print the figure object (this shows its internal representation, not the actual chart)
print(fig)

# Display the interactive line chart
fig.show()
```

```python
# Define a dictionary to represent a Plotly figure
fig = dict({
    "data": [{"type": "bar",  # Specify the type of chart (bar chart)
              "x": [1, 2, 3],  # Define the x-axis values
              "y": [1, 3, 2]}],  # Define the y-axis values (bar heights)

    # Define the layout of the figure
    "layout": {"title": {"text": "A Figure Specified By Python Dictionary"}}  # Set the title of the chart
})

# Import Plotly's I/O module to handle figure rendering
import plotly.io as pio

# Display the figure using the low-level show function
pio.show(fig)
```

```python
# Import the Plotly Graph Objects module for creating customized visualizations
import plotly.graph_objects as go

# Create a figure object using the Figure() constructor
fig = go.Figure(
    # Define the data for the figure
    data=[go.Bar(x=[1, 2, 3], y=[1, 3, 2])],  # Creates a bar chart with x-axis labels and corresponding y-values

    # Define the layout of the figure
    layout=go.Layout(
        title=go.layout.Title(text="A Figure Specified By A Graph Object")  # Set the title of the figure
    )
)

# Display the figure in an interactive visualization window
fig.show()
```

```python
# Import the Plotly Express library for creating interactive visualizations
import plotly.express as px

# Load the Iris dataset, a well-known dataset in machine learning
df = px.data.iris()  # This retrieves a built-in dataset containing measurements of iris flowers

# Create a scatter plot using Plotly Express
fig = px.scatter(
    df,  # Use the Iris dataset
    x="sepal_length",  # Define the x-axis as the sepal length
    y="sepal_width",  # Define the y-axis as the sepal width
    color="species",  # Different colors for each species of iris flower
    title="Iris Dataset Scatter Plot"  # Set the title of the plot
)

# Display the interactive scatter plot
fig.show()
```

```python
# Import the Plotly Express library for creating interactive visualizations
import plotly.express as px

# Define the categories for the x-axis
categories = ['A', 'B', 'C', 'D', 'E']

# Define the corresponding values for each category (bar heights)
values = [15, 22, 18, 12, 28]

# Create a bar chart using Plotly Express
fig = px.bar(
    x=categories,  # Set the category labels on the x-axis
    y=values,  # Set the values on the y-axis (bar heights)
    title="Simple Bar Chart"  # Set the title of the chart
)

# Display the interactive bar chart
fig.show()
```

```python
# Import the Plotly Graph Objects module for creating detailed visualizations
import plotly.graph_objects as go

# Define data points for the x-axis
x = [1.5, 2.9, 3, 4.2, 5.6]

# Define corresponding data points for the y-axis
y = [2.2, 13.3, 4.4, 55.3, 52.1]

# Create a new figure object
fig = go.Figure()

# Add a scatter plot trace to the figure
fig.add_trace(go.Scatter(
    x=x,  # Assign x-axis values
    y=y,  # Assign y-axis values
    mode='markers',  # Use markers (dots) instead of lines
    marker=dict(color='red', size=12)  # Customize marker color and size
))

# Update the figure layout with titles for clarity
fig.update_layout(
    title="Custom Scatter Plot",  # Set the chart title
    xaxis_title="X Axis",  # Label the x-axis
    yaxis_title="Y Axis"  # Label the y-axis
)

# Display the interactive scatter plot
fig.show()
```

```python
# Import the Plotly Graph Objects module for detailed visualizations
import plotly.graph_objects as go

# Define x and y values for the line chart
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 10]

# Create a figure object and add a line chart trace
fig = go.Figure(data=go.Scatter(
    x=x,  # X-axis values
    y=y,  # Y-axis values
    mode='lines'  # Use lines to connect points
))

# Update the layout to add a title and axis labels
fig.update_layout(
    title="Simple Line Chart",  # Set the chart title
    xaxis_title="X",  # Label for the x-axis
    yaxis_title="Y"  # Label for the y-axis
)

# Display the interactive line chart
fig.show()
```

```python
# Import the Plotly Express library for interactive visualizations
import plotly.express as px

# Load the Gapminder dataset and filter it to include only the year 2007
df = px.data.gapminder().query("year == 2007")

# Create a scatter plot using Plotly Express
fig = px.scatter(
    df,  # Use the filtered dataset
    x="gdpPercap",  # GDP per capita (economic factor) on the x-axis
    y="lifeExp",  # Life expectancy (health factor) on the y-axis
    size="pop",  # Bubble size represents population
    color="continent",  # Color-coded continents for easy differentiation
    hover_name="country",  # Show country names when hovering over points
    log_x=True  # Apply logarithmic scaling to GDP per capita for better visualization
)

# Display the interactive scatter plot
fig.show()
```
